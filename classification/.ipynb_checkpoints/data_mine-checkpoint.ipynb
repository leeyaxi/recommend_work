# 长春市二手房价格影响因素分析
## 摘要：

由于近年来房价出现不断攀升的趋势，楼房层出不穷，百姓买房问题日趋严重，各大城市纷纷推出抑制房价过快增长的政策。由此，想探究影响房价的因素，尤其是在鱼龙混杂的二手房市场中，房价的影响因素更为复杂。

其中，房屋自身因素主要包括房屋成新度、产权状况、楼层、朝向、户型、装修、物业类型、通风、采光、景观等，其中对房屋价格影响较大的是房屋成新度、产权状况、楼层、朝向及户型格局。

综上所述，二手房价格是由多种因素综合影响的结果，通过对以上因素的简单分析，本案例以长春二手房为研究对象，通过描述性统计分析从内部影响因素对各个区数据进行分析，我们希望能够通过回归研究经验得出比较主要的几个因素对房价的贡献，以此确定主要影响因素。通过建立模型分析研究【区域、地铁、楼层、房屋⾯积、关注度等】多个因素对⼆⼿房价的影响情况及【各个因素特征与房价】之间的关联；最终得出结论哪些因素对房价影响显著；

### 一、选题背景及问题的提出
房地产业现已成为我国国民经济的支柱产业，也成为许多地区振兴经济的重要手段。但随着房价持续上涨，使很大一部分住房需求者失去购买力；还有很大一部分消费者虽然购买了住宅，但还债压力巨大，变成了负债累累的房奴。这些都严重影响了人民的正常生活，进而影响到社会的稳定。如何在发展经济的同时平抑房价？是迫切需要科学的理论作支撑，并有效解决的现实问题。二手房市场经历着各种政策的伤痛和买卖双方及中介机构的博弈，房价成为了房产交易的敏感区，由于近年来房价出现不断攀升的趋势，百姓买房难问题日趋严峻，各中大城市纷纷推出抑制房价过快增长的政策。随着商品房房价的上涨以及二手房的各种优势,越来越多的购房者将目光转移到二手房的交易市场中。而在实际的交易过程中,一栋二手房究竟值多少钱?影响这栋二手房房价的因素到底有哪些?它们的影响程度究竟有多大?一方面,这关系到购房者的切身利益,另一方面,二手房房主也想挖掘这栋房子的最大价值。然而,人们对二手房的价格大多停留在感觉上,或者在与同类二手房各个方面的对比中来判断自己的选择是否值得,而对房屋价格的评估缺乏定量的认识

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
import os
pd.set_option("display.max_columns", None)
plt.style.use("fivethirtyeight")#fivethirtyeight是一个样式
sns.set_style({'font.sans-serif':['simhei','Arial']})
```

### 二、数据处理与说明

我们通过获取⾃链家⽹的⻓春市⼆⼿房信息，分别取2022年⻓春市⼆⼿房信息10.5号、10.19号、11.2号三个时间点的数据，总共9000条数据；

数据选取12个变量进行分析，其中3个连续型变量，9个离散型变量。因变量为单位面积房价，其他变量均为自变量。

```python
# 数据加载，三个数据集拼接

file_path = './datas/'
all_data = []
for root,dirs,files in os.walk(file_path):
    for file in files:
        tmp_df = pd.read_excel(root + file)
        all_data.append(tmp_df)
all_df = pd.concat(all_data, axis=0, ignore_index=True)
print(all_df.columns)
print(all_df.shape)
# 使用9006条数据，其中每条数据包含54个变量，
```

#### 变量分析和变量选择

```python
# 选择有效变量【因变量为房屋单价，其他都为自变量】
# 内部相关自变量：面积、卧室数、客厅数、装修、电梯、供暖方式
# 外部相关自变量：地铁、区域、关注数
df = all_df[['信息2', '关注', '关键词', '价格4', '所在区域', '供暖方式', '配备电梯1']].copy()
print(df.columns)
print(df.shape)
```

```python
print(df.tail())  # 查看数据样例
print(df.info())  # 查看数据的情况，包括数据量、缺失情况、数据类型
```

```python
# 缺失值处理
# 数据存在三条房价数据信息为空，直接删除
loss_df = df.copy()
loss_df.dropna(axis=0, how='all', subset=['信息2'], inplace=True)
loss_df['subway'] = loss_df['关键词'].apply(lambda x: 1 if '近地铁' in str(x) else 0)
loss_df.drop('关键词', axis=1, inplace=True)
loss_df['heat'] = loss_df['供暖方式'].apply(lambda x: 1 if '集中供暖' in str(x) else (2 if '自供暖' in str(x) else 0))
loss_df.drop('供暖方式', axis=1, inplace=True)
loss_df['lift'] = loss_df['配备电梯1'].apply(lambda x: 1 if '电梯' in str(x) else 0)
loss_df.drop('配备电梯1', axis=1, inplace=True)

# print(loss_df.info())
```

```python
# 数据转换
fromat_df = loss_df.copy()
fromat_df = fromat_df.rename(columns={'价格4': 'price', '关注' : 'focus_num', '所在区域': 'zone'})
fromat_df['price'] = fromat_df['price'].apply(lambda x: int(x[:-3].replace(',', '')))
fromat_df['bedroom'] = fromat_df['信息2'].apply(lambda x: int(x.split('|')[0][0]))
fromat_df['living_room'] = fromat_df['信息2'].apply(lambda x: int(x.split('|')[0][2]))
fromat_df['area'] = fromat_df['信息2'].apply(lambda x: float(x.split('|')[1][:-3]))
fromat_df['decorate'] = fromat_df['信息2'].apply(lambda x: x.split('|')[3])
fromat_df['decorate'] = fromat_df['decorate'].apply(lambda x: 1 if '简装' in str(x) else (2 if '精装' in str(x) else 0))
fromat_df = fromat_df[['focus_num', 'zone', 'subway', 'heat', 'lift', 'bedroom', 'living_room', 'area', 'decorate', 'price']]
zone_dic = {}
for i, each in enumerate(fromat_df['zone'].unique()):
    zone_dic[each] = i
print(zone_dic)
fromat_df['zone'] = fromat_df['zone'].map(zone_dic)
# 数据清洗和处理后，得到数据表
print(fromat_df.head())
fromat_df.to_excel('./数据_学号_姓名.xlsx', index=None)
```

## 三、数据分析

综上可知，通过对本案例数据的描述性分析可以得出与⼆⼿房价密切相关的可能是卧室数、客厅数、所在区域、房屋⾯积等因素。

## 四、建模分析

由于利用回归系数模型进一步研究最具影响力的变量。通过构建回归模型，对数据进⾏建模分析，计算不同⾃变量对因变量的影响程度，并实现⼆⼿房价的预测。


```python
# 建立模型进行预测分析，使用回归模型预测
train_df = ana_df.copy()

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

X = train_df.drop(['price'], axis=1)
y = train_df.loc[:, 'price']

x_train, x_val, y_train, y_val = train_test_split(X, y, test_size=0.3)
```

```python
#线性回归
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
lineR = LinearRegression()
lineR.fit(x_train, y_train)
y_pred = lineR.predict(x_val)
print('各变量的系数：\n', lineR.coef_)
print(lineR.intercept_)  #打印出模型的参数
```

```python
sum_mean=0
for i in range(len(y_val)):
    sum_mean+=(y_pred[i]-y_val.values[i])**2
sum_erro=np.sqrt(sum_mean/len(y_val))  
# calculate RMSE by hand
print ("RMSE by hand:",sum_erro)
```

这里介绍3种常用的针对线性回归的测度。

(1)平均绝对误差(Mean Absolute Error, MAE)

(2)均方误差(Mean Squared Error, MSE)

(3)均方根误差(Root Mean Squared Error, RMSE)

这里我使用RMSE。

综上分析：
1、面积、区域、关注度、是否近地铁、卧室数，五个变量对房价的影响较大【重要程度分大小，越靠前越重要】；


## 五、结论与建议
通过对⻓春市⼆⼿房数据进行统计分析，得到如下结论 

1、影响⻓春市⼆⼿房屋价格的主要因素为：

    ⾯积、卧室数、区域、关注度、地铁； 

2、⾯积和区域对房价的影响相对较⼤； 

3、'净⽉区'和 '朝阳区' 两个区域的房价相对较贵；

由于对房价的影响因素较多，在未来进一步的研究中除考虑对房价的普遍影响之外，还要考虑大环境对房价的适当影响，例如国内整体经济环境，国家政策的打压与支持等。并可适当加入其他影响因素的研究，例如某房产的宣传力度、住房周边配套设施的完善情况、住房周围文娱场所的建设等。
